### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/Visualization/Titanic_Desaster/train.csv). 

### Step 3. Assign it to a variable titanic 

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
titanic = pd.read_csv(url)
titanic.head()
titanic.shape
```

### Step 4. Set PassengerId as the index 

```python
titanic = titanic.set_index('PassengerId')
titanic.head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
titanic.groupby('Sex')[['Ticket']].count().plot(kind='pie', subplots=True);
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
titanic.head()
sns.lmplot(data=titanic, x="Fare", y="Age", hue="Sex", fit_reg=False, size=8);
```

```python
plt.scatter(titanic.Fare, titanic.Age, c=(titanic.Sex == 'male'), alpha=0.5);
titanic.plot('Fare', 'Age', kind='scatter');
```
